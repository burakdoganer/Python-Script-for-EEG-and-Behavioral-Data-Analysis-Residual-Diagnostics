```python
import pandas as pd
import numpy as np
```

```python
eeg_data = pd.read_excel("preprocessed_eeg_data_20251209_163948.xlsx")

print("Data shape: ", eeg_data.shape)
print("\nColumn names")
print(eeg_data.columns)
```

```python
eeg_segment = eeg_data[eeg_data["is-segment"] == 1]

eeg_segment.head()
```

```python
subjects_to_remove = ["s1056", "s1072", "s1113", "s1238", "s1281", "s1338"]

eeg_segment = eeg_segment[
    ~eeg_segment["subject_id"].isin(subjects_to_remove)
]
```

Cleaning the data from the first and last segments which are excluded from our analysis

```python
eeg_segment_clean = (eeg_segment.loc[eeg_segment["is-to-scrap"] != 1].reset_index(drop= True))
eeg_segment_clean
```

```python
eeg_segment_clean["is-ad"] = eeg_segment_clean["is-ad"].fillna(0).astype(int)
```

```python
eeg_segment_clean.loc[eeg_segment_clean["is-ad"] == 0, "ad_order"] = np.nan

eeg_segment_clean.loc[eeg_segment_clean["is-ad"] == 0, "is-recall"] = np.nan

eeg_segment_clean.loc[eeg_segment_clean["is-ad"] == 1, "is-recall"] = (
    eeg_segment_clean.loc[eeg_segment_clean["is-ad"] == 1, "is-recall"].fillna(0).astype(int)
    )
```

```python
eeg_segment_clean.to_excel(
    "eeg_segment_clean.xlsx",
    index=False
)
```
